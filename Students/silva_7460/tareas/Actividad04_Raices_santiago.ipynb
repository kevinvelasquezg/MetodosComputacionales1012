El método de **Regula Falsi** es una modificación al método de Bisección. Con base en los siguientes pasos, escriba una función que calcule la raiz de una función. Pruebe con $x^2 - 4 = 0$, y examine la convergencia en función de la tolerancia.


### Pasos

1. **Inicialización**:
   - Se parte de un intervalo $[a, b]$ donde $f(a)$ y $f(b)$ tienen signos opuestos.
   - Esto garantiza que hay al menos una raíz en el intervalo (por el teorema del valor intermedio).

2. **Cálculo del punto de intersección**:
   - En lugar de tomar el punto medio como en la bisección, se calcula un punto c usando la siguiente fórmula:
     
     $$c = \frac{a \cdot f(b) - b \cdot f(a)}{f(b) - f(a)}$$

   - Esta fórmula representa la intersección de la línea que conecta los puntos $(a, f(a))$ y $(b, f(b))$ con el eje $x$.

3. **Evaluación y actualización del intervalo**:
   - Se evalúa $f(c)$.
   - Si $f(c) = 0$ (o es menor que una tolerancia establecida), $c$ es la raíz buscada.
   - Si $f(c)$ tiene el mismo signo que $f(a)$, se actualiza $a = c$.
   - Si $f(c)$ tiene el mismo signo que $f(b)$, se actualiza $b = c$.

4. **Iteración**:
   - Se repiten los pasos 2 y 3 hasta que se cumpla un criterio de parada (por ejemplo, $|f(c)| < tolerancia$ o se alcance un número máximo de iteraciones).

```python
def regula_falsi(f, a, b, tolerancia=1e-12, max_iter=1000):  #f: función que toma un num real y devuelve num real, a, b, extremos del intervalo
    #evaluamos los extremos iniciales
    fa = f(a)
    fb = f(b)
    #iteramos hasta cumplir la  tolerancia o  hasta alcanzar el máximo de iteraciones
    for iteracion in range(max_iter):
        #passo 2: Cálculo del punto de intersección
        c = (a * fb - b * fa) / (fb - fa)
        fc = f(c)

        #muestra los valores actuales
        print(f"Iteración {iteracion+1}: c = {c}, f(c) = {fc}")

        #paso 3: se evalua fc
        if abs(fc) < tolerancia:
            return c  #hasta que  Se encontra la raíz con la tolerancia deseada

        #Actualiza el intervalo según el signo de fc
        if fc * fa < 0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc

#usamos la funcion que se nos brinda
def funcion(x):
    return x**2 - 4

raiz = regula_falsi(funcion, 0, 3, tolerancia=1e-12, max_iter=1000)
print(f"La raíz aproximada es: {raiz}")
```

Una carga total $Q$ se distribuye uniformemente alrededor de un conductor en forma de anillo de radio $a$. Una carga $q$ se encuentra a una distancia $x$ horizontalmente a través del centro del anillo. La fuerza ejercida sobre la carga por el anillo está dada por:

$$F = \frac{1}{4\piϵ_0}\frac{qQx}{(x^2+a^2)^{3/2}}$$

con $q$ y $Q$ de $2\times 10^{-5}$C.

  * Escriba una función donde ingresen los valores de $a$ y $x$ para hallar la fuerza.
  * Haga un gráfico de $F(x)$ para un anillo de radio $a= 0.85$m para $0.1\leq x\leq 2.0$m
  * Usando el método de Newton-Raphson y otro diferente, para encontrar el valor de $x$ en el cual la fuerza será equivalente a 1.25 N.
  * Encuentre qué sucede si disminuimos el radio con respecto a la fuerza que experimentará la carga $q$.

```python
#parametros
epsilon_0 = 8.854e-12
q = 2e-5
Q = 2e-5

## Función que calcula la fuerza F(x)
def F(x, a): #x es la posicion sobre el eje, a es el radio del anillo, la función nos retorna la Fuerza F

    return (1 / (4 * np.pi * epsilon_0)) * (q * Q * x) / ((x**2 + a**2)**(3/2))

#derivamos para poder aplicar metodo newton raphson
def dFdx(x, a):

    num = (1 / (4 * np.pi * epsilon_0)) * q * Q * ((x**2 + a**2)**(3/2) - 3*x**2*(x**2 + a**2)**(1/2))

    den = (x**2 + a**2)**3

    return num / den

#grafico
def graficar_fuerza(a):
    x_vals = np.linspace(0.1, 2.0, 200)
    f_vals = [F(x, a) for x in x_vals]

    plt.figure(figsize=(10,8))
    plt.plot(x_vals, f_vals, label=f"a = {a} m")
    plt.axhline(1.25, color='darkred', linestyle='--', label="F = 1.25 N")
    plt.xlabel("x (m)")
    plt.ylabel("Fuerza (N)")
    plt.title("Fuerza F(x) sobre una carga q debido a un anillo cargado")
    plt.grid(True, alpha = 0.4)
    plt.legend()

#método newton raphson
def newton_raphson(a, tol = 1e-12, max_iter = 1000):

    x = 0.5#valor inciial

    for i in range(max_iter):
        fx = F(x, a) - 1.25
        dfx = dFdx(x, a)
        x_nuevo = x - fx / dfx
        if abs(F(x_nuevo, a) - 1.25) < tol:

            return x_nuevo
        x = x_nuevo
    print("No converge.")
    return x

#metodo distinto: bisección
def biseccion(a, tol = 1e-12, max_iter = 1000):

    a_x = 0.1
    b_x = 2.0

    for i in range(max_iter):

        c = (a_x + b_x) / 2
        fc = F(c, a) - 1.25
        if abs(fc) < tol:
            return c
        if (F(a_x, a) - 1.25) * fc < 0:
            b_x = c
        else:
            a_x = c
    print("No converge con bisección.")
    return c

#comparamos radios
def comparar_radios():

    radios = [0.85, 0.3, 0.5, 0.6, 1] #añadimos los valores de radio a
    for r in radios:

        x_val = newton_raphson(r)
        fuerza = F(x_val, r)
        print(f"Radio a = {r} m -> x = {x_val:.11f} m, F(x) = {fuerza:.11f} N")

a = 0.85
graficar_fuerza(a)

print("Método Newton-Raphson:")
x_nr = newton_raphson(a)
print(f"x = {x_nr:.11f}, F(x) = {F(x_nr, a):.11f} N")
print("Método de bisección:")
x_bi = biseccion(a)
print(f"x = {x_bi:.11f}, F(x) = {F(x_bi, a):.11f} N")
print("Comparando radios:")
comparar_radios()
```

La velocidad ascendente $v$ de un cohete se puede calcular mediante
$$v = u\ln \frac{m_0}{m_0 - qt} - gt$$

donde $u$ es la velocidad a la que se expulsa el combustible, $m_0$ la masa inicial del cohete en $t = 0$, $q$ la razón de consumo del combustible y $g = 9.81$ m/s$^2$.

* Escriba una función para $v$ con las variables asociadas.
* Sea $m_0 = 1.6\times 10^5$ kg, $u = 1800$ m/s y $q = 2600$ kg/s realice un grafico de $v(t)$ para $0\leq t\leq 100$ s.
* Encuentre el valor de tiempo para el cual $v = 750$ m/s usando el método de Newton-Raphson, Bisección y el de Regula-Falsi
* Varie los valores de $u$ (utilice por lo menos 20) y encuentre los tiempos para los cuales el cohete alcanzará la velocidad de 750 m/s. Qué conclusiones puede sacar?



```python
from scipy.optimize import newton, bisect
#parametros
m0 = 1.6e5
u = 1800
q = 2600
g = 9.81
v_objetivo = 750

#funcion v(t) para calcilar la velocidad ascendente del cohete
def v(t, u_val = u):

    return u_val * np.log(m0 / (m0 - q * t)) - g * t

# f(t) = v(t)
def f(t, u_val = u):
    return v(t, u_val) - v_objetivo

#grafica
t_vals = np.linspace(1, 60, 1000)
v_vals = v(t_vals)

plt.figure(figsize=(10, 8))
plt.plot(t_vals, v_vals, color = 'yellow', label = 'v(t)')
plt.axhline(v_objetivo, color = 'grey', linestyle='--', label='v = 750 m/s')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Velocidad del cohete v(t)')
plt.legend()
plt.grid(True)

#newton-Raphson, con scipy es más rápido utilizar este método
t_newton = newton(f, x0 = 10) #x0 es una estimación inicial
print(f"Newton Raphson: t = {t_newton:.11f} s")

#metodo bisección
t_bisect = bisect(f, 1, 50) #intervalo de 1 a 50
print(f"Bisección: t = {t_bisect:.11f} s")

#metodo de regula falsi
def regula_falsi(func, a, b, tol = 1e-12, max_iter = 1000, u_val = u):

    fa = func(a, u_val)
    fb = func(b, u_val)

    for i in range(max_iter):

        c = b - fb * (b - a) / (fb - fa)
        fc = func(c, u_val)

        if abs(fc) < tol:
            return c

       #actualiza extremos del I
        if fa * fc < 0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc

t_regula_falsi = regula_falsi(f, 1, 50)
print(f"Regula-Falsi (manual): t = {t_regula_falsi:.11f} s")

#variamos u al menos 20 veces
u_valos = np.linspace(1000, 3000, 20)
t_results = []

for u_i in u_valos:

    try:
        #busca t tal que v = 750
        t_sol = regula_falsi(f, 1, 50, u_val = u_i)
        t_results.append((u_i, t_sol))

    except:

        t_results.append((u_i, np.nan))


print("Variación de u y tiempos correspondientes para v = 750 m/s:")

for u_i, t_i in t_results:

    if not np.isnan(t_i):
        print(f"u = {u_i:.1f} m/s -> t = {t_i:.6f} s")
    else:
        print(f"u = {u_i:.1f} m/s -> No alcanza 750 m/s")

#graficamos los resiltados
u_plot = [x for x, y in t_results if not np.isnan(y)]
t_plot = [y for x, y in t_results if not np.isnan(y)]

plt.figure(figsize=(10, 8))
plt.plot(u_plot, t_plot, marker='*', linestyle='-')
plt.xlabel('Velocidad de expulsión u (m/s)')
plt.ylabel('Tiempo para alcanzar v = 750 m/s')
plt.title('Tiempo vs Velocidad de expulsión del combustible')
plt.grid(True, alpha = 0.2)
```
